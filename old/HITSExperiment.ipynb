### Visualization

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.font_manager as font_manager
sns.set_style("darkgrid")
%pylab inline
```

```python
ustecky_authorities = pd.read_table(u"./data/Ústeckýkraj_authorities.tsv").set_index("string_id")
```

```python
fig = plt.figure()
ax = plt.subplot(111)
pl = ustecky_authorities.loc[
    ["t-mobileczechrepublic", "všeobecnázdravotnípojišťovnačeskérepubliky", 
     "telefónicaczechrepublic", "českáspořitelna", "gemoneybank",
     "cetelemčr", "essox", "proficreditczech", "smartcapital", "providentfinancial"]
].drop(
    ["name", "node_type"], axis=1
).T
pl.index = map(lambda c: int(c), pl.index)
pl.plot(ax=ax, figsize=(12, 5), fontsize=14)
ax.legend(bbox_to_anchor=(1.04, 1), loc=2, borderaxespad=0., fontsize=16)
ax.set_title("Authorities", fontsize=24)
ax.tick_params(axis='x', pad=10)
ax.tick_params(axis='y', pad=10)
```
